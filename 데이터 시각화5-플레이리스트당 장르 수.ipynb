```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train = pd.read_json("train.json", encoding='utf-8')
train
```

```python
song_meta = pd.read_json("song_meta.json", encoding="utf-8")
song_meta
```

```python
idToGenreCode = song_meta[['id','song_gn_gnr_basket']].rename(columns={"song_gn_gnr_basket":"genre_code"})
dictIdToGenreCode = idToGenreCode.set_index("id").to_dict()['genre_code']
dictIdToGenreCode
```

```python
playLists = train['songs']
```

```python
dictIdToGenreCode[21]
playLists[0]
```

```python
for playList in playLists:
    for song in playList:
        playList[playList.index(song)] = dictIdToGenreCode[song]
playLists
```

```python
import re
```

```python
p = re.compile("GN[0-9]{4}")
```

```python
p.findall(str(playLists[0]))
```

```python
p = re.compile("GN[0-9]{4}")

genreCount=[]
for playList in playLists:
    genreCount.append(len(set(p.findall(str(playList)))))

genreCount
```

```python
from collections import Counter
```

```python
count = Counter(genreCount)
genreCount = dict(count.most_common())
genreCount
```

```python
k = list(genreCount.keys())
v = list(genreCount.values())
```

```python
genreCountDf = pd.DataFrame(v,k).reset_index().rename(columns={'index':"number of genre", 0:"count"})
genreCountDf
```

```python
genreCountDf = genreCountDf.sort_values(by='number of genre')
genreCountDf
```
