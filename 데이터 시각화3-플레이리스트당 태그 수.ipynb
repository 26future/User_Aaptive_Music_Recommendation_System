```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train = pd.read_json("train.json", encoding='utf-8')
train
```

```python
tagList = list(train['tags'])
tagList
```

```python
count=[]
for tags in tagList:
    count.append(len(tags))
count
```

```python
from collections import Counter
```

```python
count = Counter(count)
tagCount = dict(count.most_common())
tagCount
```

```python
v = tagCount.values()
k = tagCount.keys()
```

```python
tagCountDf = pd.DataFrame(v,k).reset_index().rename(columns={'index':"tag", 0:'count'})
tagCountDf = tagCountDf.sort_values(by='tag')
tagCountDf
```

```python
import matplotlib
from matplotlib import font_manager, rc
import platform
if platform.system()=="Windows":
    font_name=font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
matplotlib.rcParams['axes.unicode_minus']=False

import warnings
warnings.filterwarnings("ignore")
```

```python
sns.barplot(data=tagCountDf, x='tag', y='count')
sns.set(rc={'figure.figsize':(24,18)})
plt.xlabel("태그 개수")
plt.ylabel("플레이리스트 개수")
plt.title("플레이리스트당 태그 수")
```
