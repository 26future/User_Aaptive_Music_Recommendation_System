```python
genre = pd.read_json("genre_gn_all.json", encoding='utf-8', typ='series')
genreLabel = pd.DataFrame(genre, columns=['genre']).reset_index().rename(columns={"index":"genre_code"})
dictGenreLabel = genreLabel.set_index("genre_code").to_dict()
dictGenreLabel = dictGenreLabel['genre']
dictGenreLabel
```

```python
train = pd.read_json("train.json", encoding='utf-8')
train
```

```python
idToGenreCode = song_meta[['id','song_gn_gnr_basket']].rename(columns={"song_gn_gnr_basket":"genre_code"})
dictIdToGenreCode = idToGenreCode.set_index("id").to_dict()['genre_code']
```

```python
playLists = train['songs']
playLists[0]
```

```python
# id to genre_code
for playList in playLists:
    for song in playList:
        playList[playList.index(song)] = dictIdToGenreCode[song]
playLists
```

```python
genreCodes=[]
for g in list(genreLabel['genre_code']):
    if g.endswith('00'):
        genreCodes.append(g)
genreCodes
```

```python
songsInPlaylist = list(playLists)
songsInPlaylist = str(songsInPlaylist)
```

```python
playlistCount={}
for code in genreCodes:
    playlistCount[code] = songsInPlaylist.count(code)
```

```python
#제목 (한국어)

f = open('text_data_ko.txt', 'a')

for song in songName:
    if en.search(song)!=None:
        pass
    else:   
        for p in predicates:
            for adj in adjectives:
                wspace = random.randint(1,16)
                if 1 <= wspace <= 2:
                    f.write(song+adj+'노래'+p+'\n')
                elif 2 < wspace <= 4:
                    f.write(song+adj+'노래 '+p+'\n')
                elif 4 < wspace <= 6:
                    f.write(song+adj+' 노래'+p+'\n')
                elif 6 < wspace <= 8:
                    f.write(song+" "+adj+'노래'+p+'\n')
                elif 8 < wspace <= 10:
                    f.write(song+" "+adj+' 노래'+p+'\n')
                elif 10 < wspace <= 12:
                    f.write(song+adj+' 노래 '+p+'\n')     
                elif 12 < wspace <= 14:
                    f.write(song+" "+adj+'노래 '+p+'\n')   
                elif 14 < wspace <= 16:
                    f.write(song+" "+adj+' 노래 '+p+'\n')
f.close()
```
