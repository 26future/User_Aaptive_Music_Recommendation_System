```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
song_meta = pd.read_json("song_meta.json", encoding="utf-8")
song_meta
```

```python
dateList = list(song_meta['issue_date'])
dateList
```

```python
str(dateList[0])[0:4]
```

```python
yearList=[]
for date in dateList:
    yearList.append(str(date)[0:4])
yearList
```

```python
from collections import Counter
```

```python
count = Counter(yearList)
yearCount = dict(count.most_common())
yearCount
```

```python
k = list(yearCount.keys())
v = list(yearCount.values())
```

```python
yearCountDf = pd.DataFrame(v,k).reset_index().rename(columns={'index':'year',0:'count'})
yearCountDf
```

```python
yearCountDf[yearCountDf['year']=='0']
yearCountDf.drop(27, axis=0, inplace=True)
yearCountDf[yearCountDf['year']=='0']
```

```python
import matplotlib
from matplotlib import font_manager, rc
import platform
if platform.system()=="Windows":
    font_name=font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
matplotlib.rcParams['axes.unicode_minus']=False

import warnings
warnings.filterwarnings("ignore")
```

```python
sns.barplot(data=yearCountDf, x='year', y='count')
sns.set(rc={'figure.figsize':(28,20)})
plt.xticks(rotation=48)
plt.title("발매 연도별 곡 비중")
```
